```python
class CurrencyConverter:
    def __init__(self):
        # Exchange rates for each currency, define attributes
        self.exchange_rates = {
            'CLP': 0.0011,
            'ARS': 0.0012,
            'USD': 1.0,
            'EUR': 1.10,
            'TRY': 0.034,
            'GBP': 1.27
        }

        # Minimum and maximum amounts for each currency
        self.min_amount = {'CLP': 100, 'ARS': 10, 'USD': 1, 'EUR': 1, 'TRY': 5, 'GBP': 1}
        self.max_amount = {'CLP': 1000000, 'ARS': 10000, 'USD': 1000, 'EUR': 1000, 'TRY': 5000, 'GBP': 1000}

        self.commission_rate = 0.01  # 1% commission

    def convert_currency(self, amount, from_currency, to_currency, withdraw_funds):
        if amount < self.min_amount[from_currency] or amount > self.max_amount[from_currency]:
            return "Invalid amount for the selected currency."

        if from_currency == to_currency:
            return "You are converting to the same currency. No operation needed."

        converted_amount_usd = amount * self.exchange_rates[from_currency]
        converted_amount =converted_amount_usd / self.exchange_rates[to_currency]


        if withdraw_funds:
            commission = converted_amount * self.commission_rate
            converted_amount -= commission

        return f"{amount} {from_currency} is equal to {converted_amount:.2f} {to_currency}."

def main():
    converter = CurrencyConverter()

    while True:
        print("\n===== Currency Converter =====")
        print("Available currencies: CLP, ARS, USD, EUR, TRY, GBP")

        from_currency = input("Enter your initial currency: ").upper()
        to_currency = input("Enter the currency you want to exchange to: ").upper()

        amount = float(input("Enter the amount to convert: "))

        withdraw_funds = input("Do you want to withdraw funds? (yes/no): ").lower() == 'yes'

        result = converter.convert_currency(amount, from_currency, to_currency, withdraw_funds)
        print(result)

        another_operation = input("Do you want to perform another operation? (yes/no): ").lower()
        if another_operation != 'yes':
            print("Thank you for using the currency converter. Goodbye!")
            break

if __name__ == "__main__":
    main()
```
